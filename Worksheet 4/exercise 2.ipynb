```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000 # total darts thrown

# Generate random points in the square [-1,1] x [-1,1]
x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)

# Compute distances from the center
distances = np.sqrt(x**2 + y**2)

# Count darts inside the circle
n_inside = np.sum(distances <= 1)

# Monte Carlo estimation of pi
pi_estimate = 4 * n_inside / N

print("Number of darts:", N)
print("Darts inside the circle:", n_inside)
print("Estimated value of pi:", pi_estimate)
```

```python
plt.figure()
plt.scatter(x[distances <= 1], y[distances <= 1], s=1, label="Inside circle")
plt.scatter(x[distances > 1], y[distances > 1], s=1, label="Outside circle")
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Monte Carlo Estimation of Ï€")
plt.show()
```
