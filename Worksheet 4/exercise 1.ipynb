## Choice of Random Variable (Important)

We choose a simple and well-known distribution:

### Uniform distribution on [0,1]

- True mean: $\mathbb{E}[X] = \frac{1}{2}$

This makes the convergence easy to verify visually.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
N = 10000

# Number of samples
N = 10000

# Generate random variables U ~ Uniform(0,1)
X = np.random.uniform(0, 1, N)

# Compute the running (cumulative) mean
running_mean = np.cumsum(X) / np.arange(1, N + 1)

# True mean of Uniform(0,1)
true_mean = 0.5
````

````python
# Plot
plt.figure()
plt.plot(running_mean, label="Sample mean")
plt.axhline(true_mean, linestyle='--', label="True mean (0.5)")
plt.xlabel("Number of samples (n)")
plt.ylabel("Mean value")
plt.title("Law of Large Numbers (Uniform[0,1])")
plt.legend()
plt.show()
````
